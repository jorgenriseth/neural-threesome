```python
% matplotlib inline
from matplotlib import pyplot as plt
from pathlib import Path
import Glutamate as gm
```

```python
subdom_dict = {
    "extra-cellular space": 1,
}

boundary_dict = {
    "extra-cellular border": 1,
    "pre-synaptic terminal": 2,
    "post-synaptic terminal": 3,
    "astrocyte": 4,
    "dendritic membrane": 5,
}

mesh_path = Path("./synapse_mesh/")
gl_mesh = gm.GlMesh()
gl_mesh.load_xdmf(mesh_path, subdom_dict, boundary_dict)
```

```python
fig, ax = plt.subplots()
ax.set_title('domains')
for sub_i, key in enumerate(subdom_dict):
    ax.plot(*gl_mesh.domains[sub_i].coordinates().T, ".", label=key)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_title('boundaries')
for sub_i, key in enumerate(boundary_dict):
    ax.plot(*gl_mesh.boundaries[sub_i].coordinates().T, ".", label=key)
ax.legend()
plt.show()
```
